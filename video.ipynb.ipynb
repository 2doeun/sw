1. 동영상 읽기

```python
try:
    cap = cv2.VideoCapture('video.mp4')
    print('동영상 읽기 성공')
except:
    print('동영상 읽기 실패')
```

2. 동영상 출력하기
: 초당 30프레임의 이미지로 구성

- ret: 잘 읽어왔는지 True, False로 표현
- frame: 1프레임의 이미지 저장

```python
while True:
    # 동영상에서 이미지를 프레임 단위로 읽어옴
    ret, frame = cap.read()
    
    if not ret: # ret이 False일때 (파일에 문제가 있거나 동영상이 끝났을 때)
        print('이미지 읽기 실패 또는 모두 읽음')
        # 비디오 종료
        cap.release()
        # 윈도우 실행창 종료
        cv2.destroyAllWindows() # 현재 열린 모든 창을 닫는 데 사용
        # 반복문 종료
        break
        
    # 동영상을 잘 읽어왔을때
    cv2.imshow('animation',frame)
    
    # waitKey(33):한장의 사진을 0.033초 동안 띄움
    cv2.waitKey(33)
    
    k = cv2.waitKey(33)

    # 1을 누르면 동영상 종료
    if k == 49: # ASCII코드로 1을 의미
        print('동영상 종료')
        cap.release()
        cv2.destroyAllWindows()
        break
```

# 3. 원하는 시간에 동영상 종료
- 키보드에서 1버튼을 누르면 동영상이 종료된다, 다른 버튼을 누르면 종료 안 됨!!

4. 동영상 녹화
- 동영상을 읽어온 후, 녹화를 위한 설정과 코덱을 설정해주어야 함(MPEG → mp4, DIVX → avi)
- 녹화한 영상을 저장하기 위한 설정도 필요 cv2.VideoWriter(저장할 파일명, 코덱, 프레임수, 동영상 크기)
- 키보드에서 2버튼을 눌렀을 때 녹화가 되게 조건문 생성

```python
# 녹화하기
try:
    cap = cv2.VideoCapture('video.mp4')
    print("동영상 읽기 성공")
except:
    print("동영상 읽기 실패")
    
# 녹화 설정(녹화하기 위해서는 설정 작업 필요)
fps = 30.0 # 녹화할 프레임 수
width = int(cap.get(3)) # 녹화 영상의 가로 크기
height = int(cap.get(4)) # 녹화 영상의 세로 크기

# 코덱 설정(동영상을 어떤 형식으로 압출할 것인지)
# MPEG > mp4
# DIVX > avi
fcc = cv2.VideoWriter_fourcc(*'DIVX')

# 녹화 저장 설정
# CV2.VideoWriter(저장할 파일명, 코덱, 프레임 수, 동영상 크기)
out = cv2.VideoWriter('out.avi', fcc, fps, (width, height))

record = False

while True:
    ret, frame = cap.read()
    
    if not ret:
        print("이미지 읽기 실패 또는 모두 읽음")
        cap.release()
        cv2.destroyAllWindows()
        break
        
    cv2.imshow('animation',frame)
    out.write(frame)
    
    k = cv2.waitKey(33)
    
    if k == 49:
        print("동영상 종료")
        cap.release()
        cv2.destroyAllWindows()
        break
        
    if k == 50: # 녹화 시작 버튼(2를 눌렀을 시)
        record = True
        print("녹화 시작")
        
    if k == 51:
        # 녹화 종료
        out.release()
        print("녹화 종료")
```

5. 영상 캡쳐 -> 이미지로 저장
- 여러개의 사진을 캡처하기 때문에 모두 다른 이름으로 저장하기 위해서 num 변수 생성
- 2번 버튼을 눌렀을 때 캡처하도록 조건문 생성

cv2.imwrite(저장할 파일명, 저장할 이미지, 압축설정)
cv2.IMWRITE_JPEG_PROGRESSIVE:jpg파일로 이미지 저장
0:압축률(0~9) -> 0은 압축하지 않음을 의미

```python
# 영상 캡처해서 이미지로 저장하기
# 녹화하기
try:
    cap = cv2.VideoCapture('video.mp4')
    print("동영상 읽기 성공")
except:
    print("동영상 읽기 실패")
    
num = 0 # 파일 이름 저장용

while True:
    ret,frame = cap.read()
    
    if not ret:
        print("이미지 읽기 실패 또는 모두 읽음")
        cap.release()
        cv2.destroyAllWindows()
        break
        
    cv2.imshow('animation', frame)
    out.write(frame)
    
    k == cv2.waitKey(33)
    
    if k == 49:
        print("동영상 종료")
        cap.release()
        cv2.destroyAllWindows()
        break
        
    if k == 50:
        print('./cap_img/cap_img{}.jpg 파일 저장'.format(num))
        cv2.imwrite('./cap_img/cap_img{}.jpg'.format(num),frame,
                    params = [cv2.IMWRITE_JPEG_PROGRESSIVE, 0])
        
        num += 1
```
